## Esercizio 2: Voti

Scrivere un programma che riceva in input un voto in lettere inserito dall’utente, lo traduca nel corrispondente voto numerico e infine lo stampi

- I voti in lettere sono `'A'`, `'B'`, `'C'`, `'D'` e `'F'`, eventualmente seguiti da un segno `+` o `–`.

- I loro valori numerici sono, nell’ordine, `4.0`, `3.0`, `2.0`, `1.0` e `0.0`.

- I voti `'F+'` e `'F–'` non esistono.

- Un segno `+` aumenta il voto numerico di `0.3`, mentre un segno `–` lo diminuisce della stessa quantità.

- Il voto `'A+'` è comunque uguale a `4.0`.

````python
# Scrivi qua il tuo codice

# DA PPBD02-13:

def request_voto():

    while True:

        voto_alpha = input("Inserire un voto: ")
        
        if len(voto_alpha) < 3:
           
           if len(voto_alpha) == 1  or voto_alpha[1] in ['-', '+']:
                voto_alpha = voto_alpha.lower()                         # Converto in minuscolo l'input per fare i controlli
                
                if voto_alpha[0] in ['a', 'b', 'c', 'd', 'f']:
                    return voto_alpha
                else:
                    print(f'Il carattere "{voto_alpha[0]}" non è un voto.')
                    continue

            else:
                print(f"Secondo valore non corretto: {voto_alpha}!")
                continue

        else:
            print(f"Valore troppo lungo: {voto_alpha}.")
            continue            

        
def put_Value(x):             # La funzione vuole una x, un numero

    if x[0] == 'a':
        value = 4.0
    elif x[0] == 'b':
        value = 3.0
    elif x[0] == 'c':
        value = 2.0
    elif x[0] == 'd':
        value = 1.0
    elif x[0] == 'f':
        value = 0.0
                                  # Occhio alla mancanza di else, bisogna evitare che ci siano altri casi e non ci sia l'else
    return value


def second_value(segno, voto):    # Posso rinominare la variabile e lo fa all'interno della funzione stessa, non lo sostituisce ovunque

    if segno == '-':
        voto -= 0.3
        # In caso il voto sia inizialmente 0 (F),
        # mi ritroverei un -0.3
        if voto == -0.3:
            # Riporto il voto a 0 (F)
            voto += 0.3
          
    else:            
        
        voto += 0.3
        # In caso il voto sia inizialmente 0 (F) o 4 (A),
        # mi ritroverei uno 0.3 o 4.3
        if voto in (0.3, 4.3):
            # Riporto il voto a 0 (F) o 4 (A)
            voto -= 0.3
          
    
    return voto
    
# Main()
    
voto_string = request_voto()
voto_digit = put_Value(voto_string)     # 4.0, 3.0, ...

if len(voto_string) > 1:
    voto_digit = second_value(voto_string[1], voto_digit) 

print (f'Il voto {voto_string.upper()} è equivalente a {voto_digit}.')
````
